# Test the permutations of a task

```python
from environment import *
from actions import *
from robot_task_2 import *
import matplotlib.pyplot as plt
import numpy as np
import pickle
from rules_1_1 import print_rule
```

### Initialise the environment and visualise it

```python
env=create_env()
fig,ax=plt.subplots(figsize=(14,9))
plot_env(env,ax)
```

### Let's define the task as the combination of following subtasks:
1. Pickup green and blue objects from the right bottom corner
2. Move all red and blue triangles and circles from a particular area to a desired area
3. Throw away all the green/blue squares from a target area

### Also let us read the rules and print them

```python
with open ('rules.txt', 'rb') as fp:
    rules = pickle.load(fp)
for i,rule in enumerate(rules):
    print_rule(rule,i)
```

### Let's perform the subtask1: Pickup green and blue objects
Highlighted below are the green triangles to be picked up.

```python
target_area=[position(0.25,-1),position(1.0,-0.7)]
#Plot target area
fig,ax=plt.subplots(figsize=(14,9))
plot_env(env,ax)
ax.fill_between(np.linspace(target_area[0].x,target_area[1].x,50),
                target_area[0].y,target_area[1].y,color=(57/255,62/255,68/255,0.25))
plt.legend(["Target Area"]);
```

Define the task, perform it and visualise results of last permutation. 
***Picked-up Objects are hollow***

```python
task_1=task("pickup",["g","b"],np.nan,target_area,np.nan)
my_robot=robot(task_1,env)
my_robot.make_permutations(rules,1)
fig,ax=plt.subplots(figsize=(14,9))
plot_env(env,ax)
ax.fill_between(np.linspace(target_area[0].x,target_area[1].x,50),
                target_area[0].y,target_area[1].y,color=(57/255,62/255,68/255,0.25))
plt.legend(["Target Area"]);
```

### Let's perform the subtask2: Move red/blue traingles and circles
Highlighted below are the target area and destination area

```python
fig,ax=plt.subplots(figsize=(14,9))
plot_env(env,ax)

target_area=[position(0.5,-0.75),position(1.0,0.45)]
#Plot target area
ax.fill_between(np.linspace(target_area[0].x,target_area[1].x,50),
                target_area[0].y,target_area[1].y,color=(57/255,62/255,68/255,0.25))

destination_area=[position(-1,0.4),position(-0.7,1)]
ax.fill_between(np.linspace(destination_area[0].x,destination_area[1].x,50),
                destination_area[0].y,destination_area[1].y,color=(255/255,87/255,86/255,0.2))
plt.legend(["Target Area","Destination Area"]);
```

Define the task, perform it and visualise results of last permutation.

```python
task_2=task("move",["r","b"],["triangle","circle"],target_area,destination_area)
my_robot=robot(task_2,env)
my_robot.make_permutations(rules,2)

fig,ax=plt.subplots(figsize=(14,9))
plot_env(env,ax)
ax.fill_between(np.linspace(target_area[0].x,target_area[1].x,50),
                target_area[0].y,target_area[1].y,color=(57/255,62/255,68/255,0.25))

destination_area=[position(-1,0.4),position(-0.7,1)]
ax.fill_between(np.linspace(destination_area[0].x,destination_area[1].x,50),
                destination_area[0].y,destination_area[1].y,color=(255/255,87/255,86/255,0.2))
plt.legend(["Target Area","Destination Area"]);
```

### Let's perform the subtask3: Throw away all squares
Highlighted below is the target area

```python
target_area=[position(-1,-0.4),position(-0.25,0.3)]
#Plot target area
fig,ax=plt.subplots(figsize=(14,9))
plot_env(env,ax)
ax.fill_between(np.linspace(target_area[0].x,target_area[1].x,50),
                target_area[0].y,target_area[1].y,color=(57/255,62/255,68/255,0.25))
plt.legend(["Target Area"]);
```

Perform the subtask and visualise results of last iteration

```python
task_3=task("throw",np.nan,"square",target_area,np.nan)
my_robot=robot(task_3,env)
print ("Number of Objects before deleting={}".format(str(len(env[0]))))
my_robot.make_permutations(rules,3)
print ("Number of Objects after deleting={}".format(str(len(env[0]))))
fig,ax=plt.subplots(figsize=(14,9))
plot_env(env,ax)
ax.fill_between(np.linspace(target_area[0].x,target_area[1].x,50),
                target_area[0].y,target_area[1].y,color=(57/255,62/255,68/255,0.25))
plt.legend(["Target Area"]);
```
